```python
import tensorflow
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense ,SimpleRNN ,Embedding
from keras.models import Sequential
```

```python
vocab_size = 10000
(X_train, y_train),(X_test,y_test) = imdb.load_data(num_words = vocab_size)
```

```python
print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))

# Example review (integer encoded)
print("First Review:", X_train[0])
print("Label:", y_train[0])   # 1 = Positive, 0 = Negative
```

```python
# make all reveiews some length
max_len = 200
X_train = pad_sequences(X_train,maxlen=max_len,padding='post')
X_test = pad_sequences(X_test, maxlen = max_len,padding = 'post')
print("Shape after padding:", X_train.shape)
```

```python
model = Sequential()
model.add(Embedding(input_dim= vocab_size,output_dim=32,input_length= max_len))
model.add(SimpleRNN(32))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
# 4. Compile Model
# -----------------------------
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# 6. Train Model
# -----------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)
```

```python
# 7. Evaluate Model
# -----------------------------
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", accuracy)
```
