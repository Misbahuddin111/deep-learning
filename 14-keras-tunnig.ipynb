```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt
def build_model(hp):

    model = keras.Sequential()

    # Tune number of neurons
    units = hp.Int("units", min_value=64, max_value=256, step=64)

    model.add(layers.Dense(units, activation="relu", input_shape=(8,)))

    # Tune number of hidden layers
    for i in range(hp.Int("num_layers", 1, 3)):
        model.add(layers.Dense(units, activation="relu"))

    # Tune dropout
    model.add(layers.Dropout(
        hp.Float("dropout", 0.0, 0.5, step=0.1)
    ))

    model.add(layers.Dense(10, activation="softmax"))

    # Tune optimizer
    optimizer = hp.Choice(
        "optimizer",
        ["adam", "sgd", "rmsprop"]
    )

    # Tune learning rate
    learning_rate = hp.Choice(
        "learning_rate",
        [0.01, 0.001, 0.0001]
    )

    if optimizer == "adam":
        opt = keras.optimizers.Adam(learning_rate)

    elif optimizer == "sgd":
        opt = keras.optimizers.SGD(learning_rate)

    else:
        opt = keras.optimizers.RMSprop(learning_rate)

    model.compile(
        optimizer=opt,
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

```

```python
tuner = kt.RandomSearch(
    build_model,
    objective="val_accuracy",
    max_trials=10,
    directory="tuning",
    project_name="neural_network_tuning"
)
```

```python
tuner.search(
    X_train,
    y_train,
    epochs=5,
    validation_split=0.2
)
```

```python
best_hp = tuner.get_best_hyperparameters(1)[0]

print("Best Optimizer:", best_hp.get("optimizer"))
print("Best Units:", best_hp.get("units"))
print("Best Layers:", best_hp.get("num_layers"))
print("Best Dropout:", best_hp.get("dropout"))
print("Best Learning Rate:", best_hp.get("learning_rate"))
```

```python
model = tuner.hypermodel.build(best_hp)
```

```python
model.fit(
    X_train,
    y_train,
    epochs=15,
    validation_split=0.2
)
```
